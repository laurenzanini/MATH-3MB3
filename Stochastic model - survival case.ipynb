```python
from numpy import random as rd
import statistics as stats
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def bernoulli_squirrels(p, P0, N):
    P = np.zeros(N+1)
    N_vec = np.arange(N+1)
    
    for i in N_vec:
        if (i==0):
            P[i] = P0
        else:
            P[i] = rd.binomial(P[i-1], p, 1)

            
    return P
```

```python
N = 1
P0 = 100
p = 0.32
sets = 100

results = np.zeros((N+1, sets))

times = np.arange(sets) 

for i in times:
    rd.seed(i)
    results[:, i] = bernoulli_squirrels(p, P0, N)

df_bernoulli = pd.DataFrame(results)

df_bernoulli
```

```python
mean = np.zeros(2)

for i in range(len(mean)):
        mean[i] = stats.mean(df_bernoulli.iloc[i])
        
sd = np.zeros(2)

for i in range(len(sd)):
        sd[i] = stats.stdev(df_bernoulli.iloc[i])
        
summary_bernoulli = pd.DataFrame({
    'Mean': mean,
    'Standard Deviation': sd
})

summary_bernoulli
```
